```python
from keras.models import load_model
model_imageDetect = load_model('/content/drive/MyDrive/Colab Notebooks/ImageDetect/model_imageDetect.h5') 

#thu nghiem
from keras.utils import load_img
from keras.utils import img_to_array
import matplotlib.pyplot as plt
import numpy as np

img = load_img('/content/drive/MyDrive/Colab Notebooks/AIpic/A10.JPG', target_size=(32, 32, 3))
plt.imshow(img)
img = img_to_array(img)
img = img.reshape(1, 32*32*3)
img = img.astype('float32')
img = img/255
np.argmax(model_imageDetect.predict(img), axis=-1)
```
