```python
#lay mau~
import pandas as pd
import numpy as np

time_index = pd.date_range('01/01/2023', periods = 100000, freq='30s') #tgian lay mau~ va tan so lay mau~
dataframe = pd.DataFrame(index = time_index)
dataframe['Across_Amount'] = np.random.randint(1,10,100000) #chay ngau nhien so lan truy cap
#neu co dia chi IP thi co' the thay cho random or co' ca?m bien thi doc gia tri ca?m bien len

#tong ket theo tung cot
dataframe.resample('w').sum()

#lay mau~ theo thang
dataframe.resample('M').mean()
dataframe.resample('2w').mean()
```
