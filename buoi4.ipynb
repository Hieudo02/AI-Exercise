```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
url = '1.01. Simple linear regression.csv'
data= pd.read_csv(url)
x1 = data['SAT']
y = data['GPA']
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
data.hist()
```

```python
url = 'mtcars.csv'
data = pd.read_csv(url)
data.hist() #histogram

x = data.drop(data.columns[[0,2]], axis = 1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
```

```python
model = linear_model.LinearRegression()

#hoc
x = pd.DataFrame(data['wt'])
y = data['mpg']
model.fit(x, y)
print(model.intercept_)
print(model.coef_) #coefficient: he so

#do chinh xac
model.score(x, y)

#gia tri tien doan
y_predict = model.predict(x)

#sai so
residual = data['mpg']-y_predict
residual.describe()
```

```python
#bac 3 den 10
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

url = 'mtcars.csv'
data = pd.read_csv(url)

# Remove columns 0 and 2
X = data.drop(data.columns[[0, 2]], axis=1)
y = data['mpg']

# Define the range of the predictor variable
my_range = np.arange(1.5, 5.5, 0.1)
dem_range = np.arange(2, 11)
# Loop through polynomial degrees from 2 to 10
for degree in dem_range:
    
    a_range = np.arange(1, degree + 1)

    # Create the polynomial features
    X_poly = pd.DataFrame([X['wt'] ** i for i in a_range]).T

    # Fit the model
    model = linear_model.LinearRegression()
    model.fit(X_poly, y)

    print(f'Intercept của mô hình bậc {degree}: {model.intercept_}')
    print(f'Coefficients của mô hình bậc {degree}: {model.coef_}')
    print(f'R-square của mô hình bậc {degree}: {model.score(X_poly, y)}')
    # Make predictions
    predictor = pd.DataFrame([my_range ** i for i in a_range]).T
    y_pred = model.predict(predictor)

    # Plot the data and the predicted curve
    plt.figure()
    plt.scatter(X['wt'], y, color='red')
    plt.plot(my_range, y_pred, color='blue')
    plt.title('Polynomial degree: ' + str(degree))
    plt.xlabel('wt')
    plt.ylabel('mpg')
    plt.xlim(0, 9)

plt.show()
```

```python
#50_Startups
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sas

data = pd.read_csv('50_Startups.csv').drop("State", axis =1)
data.head()
sas.pairplot(data)
corr = data.corr()
sas.heatmap(corr, annot = True)
x = data.drop('Profit', axis = True)
y = data['Profit']
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_predict = regressor.predict(x_test).round(1)

calculation = pd.DataFrame(np.c_[y_test, y_predict], columns = ['Original Profit', 'Predicted Profit']) 

print(regressor.intercept_)
print(regressor.coef_) #coefficient: he so
```
