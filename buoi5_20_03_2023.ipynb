```python
from sklearn.linear_model import Perceptron

#hang: so luong mau
#cot: so dac tinh (bao nhieu bien)
x = [[0, 0], [0,1], [1,0], [1,1]]
y = [0, 1, 1, 1]

#tao mo hinh
model = Perceptron()
#train
model.fit(x, y)
#evaluate (danh gia do chinh xac)
model.score(x, y)
```

```python
#iris
from sklearn import datasets                         #thu muc chua du lieu : init (0-9), ...
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split #chia du lieu lam 2 phan: train va test
from sklearn.metrics import accuracy_score           #do luong (measure: do chinh xac, mat mat (lost), ...)
from sklearn.preprocessing import StandardScaler     #khi data khong deu nhau => chuan hoa du lieu (scaler) (DU LIEU NAY KHONG PHAI THUC TE)

#load data
iris = datasets.load_iris()
print(iris)
```

```python
#trong iris co du lieu "data" va "target"
X = iris.data
Y = iris.target
print(X[:5])
print(Y[:5])

#chia train va test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3) #testsize = 0.3 => trainsize = 0.7
#scale du lieu
sc = StandardScaler()
#training
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
model = Perceptron(max_iter = 40, eta0 = 0.1, random_state = 0)    #so lan hoc: 40 lan
                                                                  #toc do hoc: 0.1 (cang lon sai so cang nhieu) (la so duong)
                                                                  #trang thai lay gia tri: 0
model.fit(X_train_std, Y_train)
Y_pred = model.predict(X_test_std)
print(Y_pred)
print(Y_test)
print("Do chinh xac cua mo hinh: ", accuracy_score(Y_test, Y_pred))
```

```python
import pandas as pd
import numpy as np
url = 'crabs.csv'
data = pd.read_csv(url)

data1 = data[(data['speies'] =='B')] 

#Xoa cot 'sex' cu~ di va thay the bang cot 'sex' co chi so M, F dc thay the bang 1, 0
y = data['sex']
data = data.drop(data.columns[[0]], axis = 1)
data = data.replace("M",1)
data = data.replace("F",0)
data.head()
```

```python
X = data[['fontallip', 'rearwidth', 'length', 'width', 'depth']].to_numpy()
Y = data['sex'].to_numpy()

#chia train va test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3) #testsize = 0.3 => trainsize = 0.7
#scale du lieu
sc = StandardScaler()
#training
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

model = Perceptron(max_iter = 40, eta0 = 0.1, random_state = 0)    #so lan hoc: 40 lan
                                                                  #toc do hoc: 0.1 (cang lon sai so cang nhieu) (la so duong)
                                                                  #trang thai lay gia tri: 0
model.fit(X_train_std, Y_train)
Y_pred = model.predict(X_test_std)
print(Y_pred)
print(Y_test)
print("Do chinh xac cua mo hinh: ", accuracy_score(Y_test, Y_pred))
```
